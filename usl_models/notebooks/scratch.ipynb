```python
import pathlib

from usl_models.atmo_ml import dataset
from datetime import datetime


def get_cached_sim_dates(path: pathlib.Path) -> list[(str, str)]:
    all_dates = set()
    for file in path.glob('**/met_em.d03.????-??-??_??:??:??.npz'):
        relative_path = file.relative_to(path)
        sim_name = str(relative_path.parent)
        ts = datetime.strptime(relative_path.name, dataset.EXAMPLE_FILENAME_FORMAT_NPZ)
        all_dates.add((sim_name, ts.date().strftime(dataset.DATE_FORMAT)))

    sorted(all_dates)


get_cached_sim_dates(path=pathlib.Path('data/filecache'))
```
